```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use(['science', 'notebook', 'grid'])
```

```python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.1 * np.random.randn(len(x))
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o--', color='purple', lw=0.7, ms=5)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, y, 'o--', color='purple', lw=0.8, ms=7)
```

```python
plt.figure(figsize=(8, 3))
plt.ylabel('Voltage(V)')
plt.xlabel('Time(s)')
plt.title('Voltage at 3rd Electrode')
plt.plot(x, y, '--', color='blue', lw=0.8)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, y, '--', color='blue', lw=0.8, label='Component1')
plt.ylabel('Voltage(V)')
plt.xlabel('Time(s)')
plt.title('Voltage at 3rd Electrode')
plt.legend(loc='lower left')
plt.show()
```

```python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, y, 'o', color='red', label='Data')
plt.plot(x2, y2, color='green', label='fit')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.title('Voltage at 3rd Electrode')
plt.legend(loc='lower right', fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, y, 'o', color='red', label='Data')
plt.plot(x2, y2, color='green', label='fit')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.title('Voltage at 3rd Electrode')
plt.legend(loc='lower right', fontsize=10)
plt.show()
```
